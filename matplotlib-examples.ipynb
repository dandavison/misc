```python
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = np.pi
xs = np.linspace(-2*pi, 2*pi, 200)
ys = [np.cos(x) for x in xs]
plt.plot(xs, ys)

x_ticks = [-2*pi, -pi, 0, pi, 2*pi]
x_labels = [r"$-2\pi$", r"$-\pi$", r"$0$", r"$\pi$",   r"$2\pi$"]

axes = plt.gca()  # get current axes
axes.set_xticks(x_ticks)
axes.set_xticklabels(x_labels, fontsize=13)
```
